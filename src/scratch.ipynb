```python
pdb_file = "../data/v2007/1a0q/1a0q_protein.pdb"
parser = PDBParser(QUIET=True)
structure = parser.get_structure("", pdb_file)
protein_atoms = ["C", "N", "O", "S"]
protein_atoms_dict = {atom: [] for atom in protein_atoms}
for model in structure:
            for chain in model:
                for residue in chain:
                    for atom in residue:
                        atom_id = atom.get_id()
                        for atom_type in protein_atoms:
                            if atom_type in atom_id:
                                xp, yp, zp = atom.get_coord()
                                protein_atoms_dict[atom_type].append((xp, yp, zp))
                                break
```
